# 파이썬 데이터 분석 & 인공지능 첫걸음

## 1. 주피터 노트북 (Jupyter Notebook)
- 웹 브라우저에서 실행되는 **대화형 개발 환경**
- 코드(Cell)와 설명(Markdown)을 섞어 작성 가능
- 데이터 분석, 시각화, 인공지능 실습에 최적화
- 확장자: `.ipynb`

👉 **장점**
- 코드 실행 결과를 즉시 확인 가능
- 시각화 그래프, 수식, 이미지 삽입이 쉬움
- 수업/연구 보고서 형태로 공유하기 적합

👉 **단점**
- VS Code 등 전문 IDE에 비해 **코드 자동 완성, 디버깅, 프로젝트 관리 기능이 제한적임**
- 대규모 프로젝트 개발보다는 **분석·실험용 노트북 환경**에 더 적합

설치 방법: Anaconda 배포판 설치 (파이썬, 주피터, 데이터 분석 라이브러리가 함께 설치됨)  
https://www.anaconda.com/download/success


## 2. 구글 코랩 (Google Colab)
- 구글에서 제공하는 **클라우드 기반 주피터 노트북 서비스**
- 별도의 설치 필요 없음, 구글 계정만 있으면 바로 사용 가능
- GPU/TPU 무료 지원 → 딥러닝 실습에 유리

👉 **장점**
- 온라인 환경 → 어디서든 접속 가능
- 무료 GPU 지원
- 협업(공유, 댓글 기능) 용이

👉 **단점**
- 인터넷 연결 필수
- **실행 환경(세션)이 일정 시간이 지나면 초기화됨**
- 무료 환경에서는 GPU 사용이 제한될 수 있음

## 3. 비주얼 스튜디오 코드 (VS Code)
- 마이크로소프트에서 제공하는 **범용 코드 에디터**
- 확장 기능을 설치하면 **주피터 노트북 실행** 가능
- 대규모 프로젝트 관리, 파이썬 외 다양한 언어 지원
- 설치 : https://code.visualstudio.com/download  

👉 **장점**
- 가볍고 빠름
- Git 연동, 디버깅 지원
- 주피터와 함께 사용하면 노트북 + 스크립트 파일 모두 관리 가능
- 인공지능 기반 코딩 보조 기능, 참조 등 매우 많은 기능을 지원

## 4. 어떤 걸 선택할까?
- **주피터 노트북** → 데이터 분석, 실습, 보고서 작성에 최적
- **구글 코랩** → 설치 없이 GPU 사용 + 협업 환경
- **VS Code** → 프로젝트 규모가 크고, 여러 언어/파일을 통합 관리할 때 유리
